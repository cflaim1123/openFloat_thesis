```python
import matplotlib.pyplot as plt
import numpy as np 
from matplotlib.ticker import ScalarFormatter
```

```python
def Argo(t, alpha, beta, a_0):
    A = np.e**((alpha*beta-alpha)*t)*a_0#*np.e**-a_0
    return A

# def Profiles(t, rho, alpha, beta, A_0, a_0):
#     P = (rho/(-alpha + alpha*beta))*Argo(t, alpha, beta, a_0)+A_0
#     return P

def Profiles(A, t, rho, z):
    P = rho*A*t+z
    return P
```

```python
t = np.arange(1, 365*30)
alpha = np.zeros(len(t))+ 1/10
beta = np.zeros(len(t))+ 1.0001
a_0 = np.zeros(len(t)) + 4000
A_0 = np.zeros(len(t)) + 4000
z = np.zeros(len(t)) + 0

time_under = 9 #np.tanh(t)

rho = np.zeros(len(t)) + (1/time_under)*365*(1/128)#np.sin(1/9*365*1/128*(t))+10 #
A_0 = np.zeros(len(t)) + 0

A = Argo(t, alpha, beta, a_0)
# P = Profiles(t, rho, alpha, beta, A_0, a_0)
P = Profiles(A, t, rho, z)

# print(A)
print(P)
```

```python
scalar_formatter = ScalarFormatter(useOffset=False)

fig, ax = plt.subplots(figsize=(11.5, 8))
plots = []
y2 = ax.twinx()

plots += ax.plot(t, P, c='k', label = '# profiles')
plots += y2.plot(t, A, c='k', linestyle='--',label = '# floats')
# ax.plot(t, P)


# y2.spines['right'].set_color('orange')
# y2.tick_params(axis='y', colors='orange')
y2.set_ylabel('# floats', fontsize=14)
ax.set_ylabel("# profiles (x10^7)", fontsize=14)
ax.set_xlabel("TIme (# dive cycles)", fontsize=14)
y2.tick_params(labelsize=14)
ax.tick_params(labelsize=14)

labels = [l.get_label() for l in plots]
ax.legend(plots, labels, fontsize = 14)
ax.ticklabel_format(useOffset=False)
# plt.tight_layout()
plt.show()
```
